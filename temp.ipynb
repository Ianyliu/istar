```python
import pandas as pd 
cnts = pd.read_csv('data/demo-new/cnts.tsv',sep='\t',
                   index_col=0, header=0)
locs_raw = pd.read_csv('data/demo-new/locs-raw.tsv',sep='\t',
                       index_col=0, header=0)
locs_raw
```

```python
# cnts and locs_raw indexes should be the same
# calculat the min of the length of the indexes
len_index = min(len(cnts.index), len(locs_raw.index))
# only keep the index that overlaps
common_index = cnts.index.intersection(locs_raw.index)
cnts_common = cnts.loc[common_index]
locs_raw_common = locs_raw.loc[common_index]

assert locs_raw_common.index.equals(cnts_common.index), \
    "The indexes of cnts and locs_raw do not match"
    
locs_raw_common.shape, cnts_common.shape
```

```python
locs_raw_common.to_csv('data/demo-new/locs-raw-common.tsv', sep='\t')
cnts_common.to_csv('data/demo-new/cnts-common.tsv', sep='\t')
```

```python
import numpy as np
import pandas as pd
locs_df = pd.read_csv('data/demo-new/locs-raw.tsv', sep='\t',
           index_col=0, header=0)
cnts_df = pd.read_csv('data/demo-new/cnts.tsv', sep='\t',
           index_col=0, header=0)
# check row lengths match
assert locs_df.shape[0] == cnts_df.shape[0], \
    "The number of rows in locs_df and cnts_df do not match"

# sample only half of the index
sampled_index = np.random.choice(locs_df.index, size=int(len(locs_df.index)/6), replace=False)
sampled_index = np.sort(sampled_index)
# create a new dataframe with the sampled index
locs_df_sampled = locs_df.loc[sampled_index]
cnts_df_sampled = cnts_df.loc[sampled_index]

# save the sampled dataframes
locs_df_sampled.to_csv('data/demo-new/locs-raw-sampled.tsv', sep='\t')
cnts_df_sampled.to_csv('data/demo-new/cnts-sampled.tsv', sep='\t')
```
